```python
# นำเข้าคำสั่งสำหรับการสุ่มค่า
import random
# สุ่มใส่ตัวหนังสือ Yes หรือ No ลงในช่องว่าง ของ column Ever_Married
Ever_Married_Random = ["Yes", "No"]
df['Ever_Married'] = df['Ever_Married'].fillna(random.choice(Ever_Married_Random))
df.head(50)
```

```python
#บอกค่าที่หายไปในแต่ละcolumn
df.isnull().sum()
```

```python
# แปลงข้อมูล Ever_Married จาก Yes/No เป็น 1/0
df.replace(to_replace=['Yes', 'No'], value=[1, 0])
```

```python
import random
# สุ่มใส่ตัวหนังสือ Yes หรือ No ลงในช่องว่าง ของ column Graduated
Graduated_Random = ["Yes", "No"]
df['Graduated'] = df['Graduated'].fillna(random.choice(Graduated_Random))
df.head(30)
```

```python
#บอกค่าที่หายไปในแต่ละcolumn
df.isnull().sum()
```
