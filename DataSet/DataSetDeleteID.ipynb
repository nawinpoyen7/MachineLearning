```python
import numpy as np 
import pandas as pd 
import seaborn as sns  
import matplotlib.pyplot as plt 
```

```python
url='https://raw.githubusercontent.com/nawinpoyen7/MachineLearning/main/DataSet/DataChange.csv'
df=pd.read_csv(url)
df.head(5)
```

```python
# แปลง "A", "B", "C", "D" ของ Segmentation เป็น 0 1 2 3 
df.replace({'A':0},inplace=True)
df.replace({'B':1},inplace=True)
df.replace({'C':2},inplace=True)
df.replace({'D':3},inplace=True)
df.head(5)
```

```python
df.groupby('Segmentation').mean()
```

```python
df.groupby('Segmentation').max()
```

```python
df.groupby('Segmentation').min()
```

```python
df.drop(['ID'], axis=1, inplace=True)
```

```python
df.head(5)
```

```python
# save ข้อมูลทั้งหมดให้เป็นประเภท .csv และ .xls
df.to_csv('DataSetDeleteID.csv', index=False)
```
