```python
import pandas as pd
from sklearn import preprocessing
DataSet = pd.read_csv('https://raw.githubusercontent.com/nawinpoyen7/MachineLearning/main/Data%20Norm.csv')
norm_data = preprocessing.normalize(DataSet, axis=0)
norm_df = pd.DataFrame(norm_data,columns=[DataSet.columns])
print('Original Data \n',DataSet.head(10))
print('Normalized Data \n',norm_df.head(10))
```
