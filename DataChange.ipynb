```python
# นำเข้าคำสั่งสำหรับการสุ่มค่า
import random
# สุ่มใส่ตัวหนังสือ Yes หรือ No ลงในช่องว่าง ของ column Ever_Married
Ever_Married_Random = ["Yes", "No"]
df['Ever_Married'] = df['Ever_Married'].fillna(random.choice(Ever_Married_Random))
df.head(30)
```

```python
import random
# สุ่มใส่ตัวหนังสือ Yes หรือ No ลงในช่องว่าง ของ column Graduated
Graduated_Random = ["Yes", "No"]
df['Graduated'] = df['Graduated'].fillna(random.choice(Graduated_Random))
df.head(30)
```

```python
#บอกค่าที่หายไปในแต่ละcolumn
df.isnull().sum()
```

```python
import random
# สุ่มใส่ตัวหนังสือ Cat_1 ถึง Cat_7 ลงในช่องว่าง ของ column Profession
Var_1_Random = ["Cat_1", "Cat_2", "Cat_3", "Cat_4", "Cat_5", "Cat_6", "Cat_7"]
df['Var_1'] = df['Var_1'].fillna(random.choice(Var_1_Random))
df.head(30)
```

```python
#บอกค่าที่หายไปในแต่ละcolumn
df.isnull().sum()
```

```python
# เช็คประเภทข้อมุล
df.dtypes
```

```python
#แปลงประเภทข้อมูล
df['Family_Size'] = df['Family_Size'].astype('int64')
df['Work_Experience'] = df['Work_Experience'].astype('int64')
df.head(30)
```

```python
# เช็คประเภทข้อมุล
df.dtypes
```

```python
# แปลง 'High', 'Low', 'Average' ของ Spending_Score เป็น 2 1 0
df.replace({'High':2},inplace=True)
df.replace({'Average':1},inplace=True)
df.replace({'Low':0},inplace=True)
df.head(30)
```

```python
# แปลง 'Yes', 'No' ของ Ever_Married เป็น 1, 0
df.replace({'Yes':1},inplace=True)
df.replace({'No':0},inplace=True)
df.head(30)
```

```python
# แปลง 'Male', 'Female' ของ Gender เป็น 1, 0
df.replace({'Male':1},inplace=True)
df.replace({'Female':0},inplace=True)
df.head(30)
```

```python
# แปลง "Artist", "Healthcare", "Entertainment", "Engineer", 
#       "Doctor", "Lawyer", "Executive", "Marketing", "Homemaker" 
#       ของ Profession เป็น 0 1 2 3 4 5 6 7 8
df.replace({'Artist':0},inplace=True)
df.replace({'Healthcare':1},inplace=True)
df.replace({'Entertainment':2},inplace=True)
df.replace({'Engineer':3},inplace=True)
df.replace({'Doctor':4},inplace=True)
df.replace({'Lawyer':5},inplace=True)
df.replace({'Executive':6},inplace=True)
df.replace({'Marketing':7},inplace=True)
df.replace({'Homemaker':8},inplace=True)
df.sample(30)
```

```python
# แปลง "Cat_1", "Cat_2", "Cat_3", "Cat_4", "Cat_5", "Cat_6", "Cat_7" 
#       ของ Profession เป็น 0 1 2 3 4 5 6
df.replace({'Cat_1':0},inplace=True)
df.replace({'Cat_2':1},inplace=True)
df.replace({'Cat_3':2},inplace=True)
df.replace({'Cat_4':3},inplace=True)
df.replace({'Cat_5':4},inplace=True)
df.replace({'Cat_6':5},inplace=True)
df.replace({'Cat_7':6},inplace=True)
df.head(30)
```
