```python
import random
# สุ่มใส่ตัวหนังสือ Yes หรือ No ลงในช่องว่าง ของ column Profession
Var_1_Random = ["Cat_1", "Cat_2", "Cat_3", "Cat_4", "Cat_5", "Cat_6", "Cat_7"]
df['Var_1'] = df['Var_1'].fillna(random.choice(Var_1_Random))
df.head(30)
```

```python
# เช็คประเภทข้อมุล
df.dtypes
```

```python
#แปลงประเภทข้อมูล
df['Family_Size'] = df['Family_Size'].astype('int64')
df['Work_Experience'] = df['Work_Experience'].astype('int64')
```

```python
# เช็คประเภทข้อมุล
df.dtypes
```

```python
# แปลง 'High', 'Low', 'Average' ของ Spending_Score เป็น 2 1 0 และ 'Yes', 'No' เป็น '1', '0'
df.replace({'High':2},inplace=True)
df.replace({'Average':1},inplace=True)
df.replace({'Low':0},inplace=True)
df.replace({'Yes':1},inplace=True)
df.replace({'No':0},inplace=True)
```

```python
# save ข้อมูลทั้งหมดให้เป็นประเภท .csv และ .xls
df.to_csv('Data Change.csv', index=False)
df.to_excel('Data Change.xls', index=False)
```
